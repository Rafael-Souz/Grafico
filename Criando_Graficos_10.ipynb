```python
# Atividade para esquentar os motores (10 minutinhos)
# Criar uma função que permita mostrar um gráfico, a partir da configuração de alguns parâmetros.
def grafico(start, end, titulo, eixo_x, eixo_y, arquivo):
  x = np.linspace(start, end, 100)

  plt.plot(x, x, label='Linear')
  plt.plot(x, x**2, label='Quadrática')
  plt.plot(x, x**3, label='Cúbica')

  plt.title(titulo, size=18)
  plt.xlabel(eixo_x, size=14, style='italic')
  plt.ylabel(eixo_y, size=14, style='italic')
  plt.legend()

  plt.grid(True)

  plt.savefig(arquivo + '.png', dpi=100)
  plt.show()
```

```python
grafico(0, 3, 'Outro Gráfico feito com função', 'Batatinha', 'Maça', 'arquivo_02')
```
